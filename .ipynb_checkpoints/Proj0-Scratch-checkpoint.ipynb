```python
import json
import torch
```

```python
cleaned_data = []
with open("pdtb/train.json", 'r') as f:
    for line in f:
        line = json.loads(line)
        arg1 = line['Arg1']['RawText']
        conn = line['Connective']['RawText']
        arg2 = line['Arg2']['RawText']
        sense = line['Sense']
        cleaned_data.append([arg1, conn, arg2, sense])
```

```python
def conn_sense_correspondences(c:str) -> list[tuple[int, str]]:
    '''
    given string c for connective, searches the data for each sense corresponding
    to the connective and returns an enumerated list of senses s1...sn
    '''
   
    # create empty set of senses
    sense_set = set()
    # define where to look for c and s
    connectives = [instance[1] for instance in cleaned_data]
    senses = [instance[3] for instance in cleaned_data]
    # search data and add senses to set
    for i in range(len(connectives)):
        if connectives[i] == c:
            sense_set.add(tuple(senses[i]))
    # convert set to enumerated table
    return [(i, sense) for i, sense in enumerate(sense_set)]
```
